## FIFA Wolrd Cup 2018 - Luka Modric

O ano de 2018 foi brilhante para Luka Modric, craque do Real Madrid e da seleção croata. Além de ter sido eleito o melhor Jogador da Copa do Mundo FIFA, foi também consagrado com os prêmios Bola de Ouro (France Football) e FIFA The Best. No âmbito do coletivo, chegou com a Croácia ao inesperado segundo lugar na Copa do Mundo. Já com o Real Madrid, foi campeão da Uefa Champions League, mais uma vez. 

A presença de Modric no setor de meio campo do Real Madrid, foi se tornando essencial com o tempo. O croata se tornou o motor técnico e tático do clube.  Reconhecido por seu estilo com visão de jogo rápida e inteligente, organização da equipe, saída de bola com giros e inversões e passes precisos. Virou uma figura quase onipresente dentro de campo, apoiando desde a defesa ao ataque.

Na seleção da Croácia não foi diferente, hoje capitão, se tornou a referência do time na criação das jogadas, sua capacidade para tomada de decisões nas armações do jogo, aliada a sua execução precisa e flexibilidade na recomposição, fizeram de Luka o pilar do esquema tático croata. Na Copa da Rússia isso ficou nítido, foi escolhido como melhor em campo em três oportunidades, marcou dois gols e deu uma assistência.  Quando jogou abaixo de seu nível técnico viu a seleção croata sofrer e levar os jogos para incríveis três prorrogações seguidas. Sua liderança e garra fizeram a diferença em campo. Mesmo chegando na final e não conquistando a Copa do Mundo, a campanha da seleção foi excepcional e mostrou ao mundo a importância de seu camisa dez.

Seu jogo é simples e discreto, se movimenta com intensidade e oferece suporte para os demais. Poucos jogadores conseguem influenciar tanto em um esquema tático apenas com seus passes e visão das jogadas, Modric demonstra com seu talento e influência em campo que é um desses raros jogadores.

Nesta análise, vamos descrever o comportamento de Modric dentro de campo nos jogos válidos pela Copa do Mundo FIFA 2018. As ações esperadas pelo jogador, são de intensa movimentação, principalmente no meio de campo, seguido de inúmeros passes distribuídos ao longo do jogo, na qual conferem a seleção croata uma posse de bola qualificada e uma boa transição entre a defesa e o ataque.

```python
# Importando o módulo json
import json
```

```python
# Carregando os dados Copa do Mundo FIFA 2018
with open('Statsbomb/data/matches/43/3.json') as file:
    world_cup = json.load(file)
```

```python
# Primeiro componente do dado
world_cup[0]
```

```python
# Lista de partidas da Croácia com seus respectivos ID
croatia = 'Croatia'
result_croatia = []
match_id_lst = []

for match in world_cup:
    home_team_name = match['home_team']['home_team_name']
    away_team_name = match['away_team']['away_team_name']
    
    if (home_team_name == croatia) or (away_team_name == croatia):
        home_score = match['home_score']
        away_score = match['away_score']
        match_id = match['match_id']
        match_id_lst.append(match_id)
        competition_stage = match['competition_stage']['name']
        result = ('{} {} x {} {}'.format(home_team_name, home_score, away_score, away_team_name))
        result_croatia.append(result)
        print(result, match_id, competition_stage)
```

```python
# ID de cada partida da Corácia
match_id_croatia_denmark = match_id_lst[0]
match_id_croatia_nigeria = match_id_lst[1]
match_id_croatia_russia = match_id_lst[2]
match_id_croatia_argentina = match_id_lst[3]
match_id_croatia_france = match_id_lst[4]
match_id_croatia_iceland = match_id_lst[5]
match_id_croatia_england = match_id_lst[6]
```

### Campo de Jogo

```python
# Bibliotecas para construir o campo de futebol
from FCPython import createPitch 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Dimensões do campo
pitchLengthX = 120
pitchWidthY = 80
```

### Round 1 - Croatia 2 x 0 Nigeria

```python
# Acessando o jogo
file_name = str(match_id_croatia_nigeria) + '.json'

with open('Statsbomb/data/events/' + file_name, encoding='utf8') as data_file:
    data = json.load(data_file)
```

```python
# Acessando um evento qualquer da partida
data[50]
```

```python
# Transformando o arquivo data em um Dataframe
from pandas.io.json import json_normalize
df = json_normalize(data, sep = "_").assign(match_id = file_name[:-5])
```

```python
# Visualizando as primeiras linhas do Dataframe
df.head()
```

```python
# Shape do dado
df.shape
```

```python
# Verificando todas as colunas do DataFrame
for i in df:
    print(i)
```

```python
# Tipos de eventos no jogo
df['type_name'].value_counts()
```

```python
# Passes do jogo
pass_game = df.loc[df['type_name'] == 'Pass'].set_index('id')
pass_game.shape
```

```python
# Plotando os passes do Luka Modric
name_player1 = 'Luka Modrić'
count_pass_complete = 0
count_pass_incomplete = 0

# Construindo o campo
(fig,ax) = createPitch(pitchLengthX, pitchWidthY, 'yards', 'grey')

# Passes
for i, thepass in pass_game.iterrows(): 

# Posição dos passes
    if (thepass['player_name'] == name_player1) and (thepass['play_pattern_name'] == 'Regular Play'):
        if (thepass['pass_outcome_name'] == 'Incomplete') or (thepass['pass_outcome_name'] == 'Out'):
            # Posição
            x = thepass['location'][0]
            y = thepass['location'][1]
            passCircle = plt.Circle((x, pitchWidthY-y), 1, color = 'red')      
            passCircle.set_alpha(0.4)   
            ax.add_patch(passCircle)

            # Direção
            dx = thepass['pass_end_location'][0]-x
            dy = thepass['pass_end_location'][1]-y
            passArrow = plt.Arrow(x, pitchWidthY-y, dx, -dy, width = 1, color = 'red')
            passArrow.set_alpha(0.4)
            ax.add_patch(passArrow)
            
            # Contagem de passes incompletos
            count_pass_incomplete = count_pass_incomplete + 1
        else:
            # Posição
            x = thepass['location'][0]
            y = thepass['location'][1]
            passCircle = plt.Circle((x, pitchWidthY-y), 1, color = 'blue')      
            passCircle.set_alpha(1)   
            ax.add_patch(passCircle)

            # Direção
            dx = thepass['pass_end_location'][0]-x
            dy = thepass['pass_end_location'][1]-y
            passArrow = plt.Arrow(x, pitchWidthY-y, dx, -dy, width = 1, color = 'blue')
            passArrow.set_alpha(1)
            ax.add_patch(passArrow)
            
            # Contagem de passes completos
            count_pass_complete = count_pass_complete + 1
            
# Total de passes
total_pass = count_pass_complete + count_pass_incomplete

# Títulos e legendas
plt.text(5, 75, 'Passes completos', color = 'blue')
plt.text(5, 73, 'Passes incompletos', color = 'red')

fig.text(0.16, 0.85, 'Luka Modric\nCopa do Mundo 2018 - 1º Rodada\nCroácia 2 x 0 Nigéria', size = 12)
fig.text(0.16, 0.13, 'Criado por Otavio Santos / otaviosanluz@gmail.com. Dados: statsbomb.com.',
         fontstyle = 'italic')

fig.set_size_inches(12, 8)
plt.show()
```

```python
# Contagem de passes Luka Modric
percent_pass_complete = (count_pass_complete * 100) / total_pass
print('Contagem de passes Luka Modric')
print('Passes incompletos: ', count_pass_incomplete)
print('Passes completos: ', count_pass_complete)
print('Porcentagem de passes corretos: ', (round(percent_pass_complete, 2)), '%')
```

```python
# Posicionamento de Luka Modric no momento em que recebe a bola
ball_receipt_game = df.loc[df['type_name'] == 'Ball Receipt*'].set_index('id')

# Construindo o campo
(fig,ax) = createPitch(pitchLengthX, pitchWidthY, 'yards', 'grey')

# Recepção de bola
for i, ball_receipt in ball_receipt_game.iterrows(): 
    
    if (ball_receipt['player_name'] == name_player1)  and (ball_receipt['play_pattern_name'] == 'Regular Play'):
            # Posição
            x = ball_receipt['location'][0]
            y = ball_receipt['location'][1]
            passCircle = plt.Circle((x, pitchWidthY-y), 1, color = 'red', zorder = 10)      
            passCircle.set_alpha(0.8)   
            ax.add_patch(passCircle)
            
fig.text(0.16, 0.85, 'Luka Modric\nCopa do Mundo 2018 - 1º Rodada\nCroácia 2 x 0 Nigéria', size = 8)
plt.text(5, 75, 'Posicionamento ao receber a bola', color = 'red', size = 5)
fig.text(0.16, 0.13, 'Criado por Otavio Santos / otaviosanluz@gmail.com. Dados: statsbomb.com.',
         fontstyle = 'italic', size = 5)

plt.savefig('figures/ball_receipt_modric_croatia_nigeria.pdf')

fig.set_size_inches(12, 8)
plt.show()
```
