```python
# Importando bibliotecas
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
# Scraping xG 
base_url = 'https://understat.com/match/'
partida = str(input('Informe o ID da partida: '))
url = base_url + partida
```

```python
url
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')
```

```python
scripts
```

```python
# Transformando os dados para o formato string
strings = scripts[1].string
```

```python
strings
```

```python
# Dados json
inicio = strings.index("('") + 2
fim= strings.index("')")

json_data = strings[inicio:fim]
json_data = json_data.encode('utf8').decode('unicode_escape')

# Convertendo o formato string para o formato json
data = json.loads(json_data)
```
